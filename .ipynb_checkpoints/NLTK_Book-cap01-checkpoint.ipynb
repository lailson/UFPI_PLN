## 1.4 Contando vocabulário

O fato mais óbvio sobre textos que emerge dos exemplos anteriores é que eles diferem no vocabulário que usam. Nesta seção, veremos como usar o computador para contar as palavras em um texto de várias maneiras úteis.

Vamos começar descobrindo o comprimento de um texto do início ao fim, em termos das palavras e símbolos de pontuação que aparecem. Usamos o termo len para obter o comprimento de algo, que aplicaremos aqui ao livro de Gênesis:

```python
len(text3)
```

Então Gênesis tem 44.764 palavras e símbolos de pontuação, ou "tokens". Um token é o nome técnico para uma sequência de caracteres — como hairy, his ou :) — que queremos tratar como um grupo. Quando contamos o número de tokens em um texto, digamos, a frase to be or not to be, estamos contando ocorrências dessas sequências. Assim, em nossa frase de exemplo, há duas ocorrências de to, duas de be e uma de or e not. Mas há apenas quatro itens de vocabulário distintos nesta frase. Quantas palavras distintas o livro de Gênesis contém? Para resolver isso em Python, temos que colocar a questão de forma ligeiramente diferente. O vocabulário de um texto é apenas o conjunto de tokens que ele usa, já que em um conjunto, todas as duplicatas são recolhidas juntas. Em Python, podemos obter os itens de vocabulário de text3 com o comando: set(text3).

```python
sorted(set(text3))[:20] # Mostrando os 20 primeiros tokens
```

```python
len(set(text3)) #Total
```

Ao envolver sorted() em torno da expressão Python set(text3), obtemos uma lista ordenada de itens de vocabulário, começando com vários símbolos de pontuação e continuando com palavras que começam com A. Todas as palavras em maiúsculas precedem palavras em minúsculas. Descobrimos o tamanho do vocabulário indiretamente, perguntando sobre o número de itens no conjunto, e novamente podemos usar len para obter esse número [2]. Embora tenha 44.764 tokens, este livro tem apenas 2.789 palavras distintas, ou "tipos de palavras". Um tipo de palavra é a forma ou grafia da palavra independentemente de suas ocorrências específicas em um texto — ou seja, a palavra considerada como um item único de vocabulário. Nossa contagem de 2.789 itens incluirá símbolos de pontuação, então geralmente chamaremos esses itens únicos de tipos em vez de tipos de palavras.

Agora, vamos calcular uma medida da riqueza lexical do texto. O próximo exemplo nos mostra que o número de palavras distintas é apenas 6% do número total de palavras, ou equivalentemente que cada palavra é usada 16 vezes em média

```python
len(set(text3)) / len(text3)
```

Em seguida, vamos focar em palavras específicas. Podemos contar com que frequência uma palavra ocorre em um texto e calcular qual porcentagem do texto é ocupada por uma palavra específica:

```python
text3.count("smote")
```

```python
100 * text4.count('a') / len(text4)
```

Você pode querer repetir esses cálculos em vários textos, mas é tedioso ficar digitando a fórmula novamente. Em vez disso, você pode criar seu próprio nome para uma tarefa, como "diversidade_lexical" ou "porcentagem", e associá-lo a um bloco de código. 

```python
def diversidade_lexical(text): 
    return len(set(text)) / len(text) 
```

```python
def porcentagem(count, total):
    return 100 * count / total
```

Depois que o Python sabe que diversidade_lexical() e porcentagem() são os nomes de blocos específicos de código, podemos prosseguir e usar estas funções:

```python
diversidade_lexical(text3)
```

```python
diversidade_lexical(text5)
```

```python
porcentagem(4, 5)
```

```python
porcentagem(text4.count('a'), len(text4))
```

Para recapitular, usamos ou chamamos uma função como diversidade_lexical() digitando seu nome, seguido por um parêntesis de abertura, o nome do texto e, em seguida, um parêntesis de fechamento. Esses parênteses aparecerão com frequência; sua função é separar o nome de uma tarefa — como diversidade_lexical() — dos dados nos quais a tarefa deve ser executada — como text3. O valor de dados que colocamos entre parênteses quando chamamos uma função é um argumento para a função.

## 2.2 Indexando lista

Como vimos, um texto em Python é uma lista de palavras, representadas usando uma combinação de colchetes e aspas. Assim como em uma página comum de texto, podemos contar o número total de palavras em text1 com len(text1), e contar as ocorrências em um texto de uma palavra específica — digamos, 'heaven' — usando text1.count('heaven').

Com alguma paciência, podemos escolher a 1ª, 173ª ou até mesmo a 14.278ª palavra em um texto impresso. Analogamente, podemos identificar os elementos de uma lista Python pela ordem de ocorrência na lista. O número que representa essa posição é o índice do item. Instruímos o Python a nos mostrar o item que ocorre em um índice como 173 em um texto escrevendo o nome do texto seguido pelo índice dentro de colchetes:

```python
text4[173]
```

Podemos fazer o inverso; dada uma palavra, encontre o índice de quando ela ocorre pela primeira vez:

```python
text4.index('awaken')
```

Índices são uma maneira comum de acessar as palavras de um texto, ou, mais genericamente, os elementos de qualquer lista. Python nos permite acessar sublistas também, extraindo pedaços gerenciáveis de linguagem de textos grandes, uma técnica conhecida como slicing.

```python
text5[16715:16735]
```

```python
text6[1600:1625]
```
